## Formula para calcular el area
$A=\pi*r*^2$
## Formula para calcular perimetro
$P=2*\pi*r$
## Datos de entrada
$r$ ?
---
## Diagrama de flujo en LucidChart
![DF LuciChart](https://raw.githubusercontent.com/zavalaalexia128/EPPL-Area-y-perimetro-de-un-circulo-mis-imagenes/fc5e291ce00558879cf2dc79113862f5b9cab668/DF%20AreaCirculo.PNG)

```python
# Calcular Area y perimetro de cirulo
# Variable A, P, r
# Formulas A = pi*^2 y P = 2 * pi * r
# Captura del valor de r
import math
r = float(input("Ingrese el radio del circulo: "))
A = math.pi * r ** 2
P = 2 * math.pi * r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Programa realiado por Zavala Medina NL: 54")
```
